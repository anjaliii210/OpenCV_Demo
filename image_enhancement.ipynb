#### image thresholding

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img= cv.imread('alien.jpg',0)
plt.imshow(img)
```

```python
retval,img_th1=cv.threshold(img,thresh=150,maxval=255,type=cv.THRESH_BINARY)
plt.imshow(img_th1)
```

```python
retval,img_th1=cv.threshold(img,thresh=150,maxval=255,type=cv.THRESH_BINARY_INV)
plt.imshow(img_th1)
```

```python
retval,img_th1=cv.threshold(img,thresh=150,maxval=170,type=cv.THRESH_BINARY)
plt.imshow(img_th1)
```

```python
retval,img_th1=cv.threshold(img,thresh=150,maxval=255,type=cv.THRESH_TRUNC)
plt.imshow(img_th1)
```
